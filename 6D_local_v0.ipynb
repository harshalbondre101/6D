# Break down dataset [opt]
Input: takes raw data as input

Output: create small chunks for feature agent.

LLM: none

```python
# Part 1: Breaking the file into chunks using sliding window

def read_and_chunk_file_with_window(file_path, chunk_size, step_size):
    # Open the file and read its contents
    with open(file_path, 'r') as file:
        content = file.read()

    # Create chunks using a sliding window approach
    input_components = []
    for i in range(0, len(content) - chunk_size + 1, step_size):
        input_components.append(content[i:i + chunk_size])

    # Handle the last chunk if there's leftover content
    if len(content) % step_size != 0:
        input_components.append(content[-chunk_size:])

    return input_components


def save_chunks_to_file(chunks, output_file):
    # Open the output file to write chunks
    with open(output_file, 'w') as file:
        for idx, chunk in enumerate(chunks):
            file.write(f"Chunk {idx + 1}:\n{chunk}\n\n")


# Input file path, chunk size, and sliding step size
file_path = 'input/input.txt'  # Replace with the actual file path
chunk_size = 720  # Adjust the chunk size
step_size = 50  # Adjust the step size

# Get the chunks with the sliding window
input_components = read_and_chunk_file_with_window(file_path, chunk_size, step_size)

# Save the chunks to parts.txt
output_file = 'input/parts.txt'
save_chunks_to_file(input_components, output_file)

print(f"Chunks have been saved to {output_file}.")
```

```python
feature_agent = ReActAgent.from_tools(
    llm=llm3,
    verbose=True,
)

# Correcting the prompt template
react_system_header_str = """You are a feature development agent. Develop detailed features of an app based on given data. Your Output Should be JSON feature, feature_description. Do not generate anything else. No explaination needed.
"""
react_system_prompt = PromptTemplate(react_system_header_str)

feature_agent.update_prompts({"agent_worker:system_prompt": react_system_prompt})
prompt_dict = feature_agent.get_prompts()
for k, v in prompt_dict.items():
    print(f"Prompt: {k}\n\nValue: {v.template}")
```

```python
import os
import re
import concurrent.futures

# Directory setup for storing output files
output_dir = "Features"
os.makedirs(output_dir, exist_ok=True)

# Read content from the file parts.txt
file_path = "input/parts.txt"  # Path to the parts file

# Function to parse the file and extract chunks
def read_chunks_from_file(file_path):
    try:
        with open(file_path, "r", encoding="utf-8") as file:
            content = file.read()  # Read the entire file content
        # Using regular expressions to split the content by chunk headers
        chunks = re.split(r"Chunk \d+:", content)  # Split by "Chunk n:"
        # Remove any empty entries resulting from the split
        chunks = [chunk.strip() for chunk in chunks if chunk.strip()]
        return chunks
    except FileNotFoundError:
        print(f"The file at {file_path} was not found.")
        return []

# Get the chunks from the file
content = read_chunks_from_file(file_path)

# Method to generate and save user stories for each chunk
def generate_and_save_feature_list(i, feature, output_dir):
    # Assuming 'feature_agent' is an instance of ReActAgent
    user_story = feature_agent.chat(f"Develop features from given data: {feature}")

    # Define a safe filename (replace spaces & special characters)
    safe_filename = f"feature_list_{i}.txt"

    # Save the user story
    with open(os.path.join(output_dir, safe_filename), "w", encoding="utf-8") as file:
        file.write(str(user_story))  # Convert to string if needed

    print(f"Saved: {safe_filename}")

# Check if content exists before processing
if content:
    # Use ThreadPoolExecutor for parallel processing
    with concurrent.futures.ThreadPoolExecutor() as executor:
        futures = [
            executor.submit(generate_and_save_feature_list, i, feature, output_dir)
            for i, feature in enumerate(content, start=1)
        ]

        # Wait for all tasks to complete
        for future in concurrent.futures.as_completed(futures):
            future.result()  # Will raise any exception that occurred during execution
else:
    print("No content to process.")

# Print the list to verify
print(content)
```

# User Flow Developement agent
used to develop user flow from user story.

Input: user story

Output: user flow

LLM: smollm2:360m

```python
user_flow_agent = ReActAgent.from_tools(
    llm=llm6,
    verbose=True,
)

react_system_header_str = """ You are a user flow development agent.
"""
react_system_prompt = PromptTemplate(react_system_header_str)

user_flow_agent.update_prompts({"agent_worker:system_prompt": react_system_prompt})

prompt_dict = user_flow_agent.get_prompts()
for k, v in prompt_dict.items():
    print(f"Prompt: {k}\n\nValue: {v.template}")
```

```python
import os
import concurrent.futures
import logging

# Configure logging
logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(message)s")

# Define the directories
user_story_dir = "user_stories"
output_dir = "user_flows"

# Ensure the input and output directories exist
if not os.path.exists(user_story_dir):
    logging.error(f"The directory {user_story_dir} does not exist.")
    user_stories = []
else:
    # Get a list of all files in the input directory
    file_list = [
        os.path.join(user_story_dir, file)
        for file in os.listdir(user_story_dir)
        if os.path.isfile(os.path.join(user_story_dir, file))
    ]

    # Read the content of each file and store it in an array
    user_stories = []
    for file_path in file_list:
        with open(file_path, "r", encoding="utf-8") as file:
            content = file.read().strip()
            if content:  # Skip empty files
                user_stories.append((file_path, content))  # Store file path for better logging

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

def generate_and_save_user_flows(file_path, i, feature, output_dir):
    try:
        # Assuming 'user_flow_agent' is defined elsewhere in your code
        user_flow = user_flow_agent.chat(f"Develop user flow for given user story. Provide detailed user flow. Add steps and substeps. Make it as actionable as possible. Your output should be actionable numbered flow. Do not write anything else. Here is the user story: {feature}")

        # Define a safe filename for the user flow
        safe_filename = f"user_flow_{i}.txt"

        # Save the user flow in the output directory
        with open(os.path.join(output_dir, safe_filename), "w", encoding="utf-8") as file:
            file.write(str(user_flow))  # Convert to string if needed

        logging.info(f"Successfully processed: {file_path} -> {safe_filename}")
    except Exception as e:
        # Log the error and skip the file
        logging.error(f"Error processing {file_path}: {e}")
        return False  # Return False to indicate failure
    return True  # Return True to indicate success

# Check that user stories are loaded correctly
if user_stories:
    # Use ThreadPoolExecutor to run tasks in parallel
    with concurrent.futures.ThreadPoolExecutor() as executor:
        # Submit tasks to the executor
        futures = {
            executor.submit(generate_and_save_user_flows, file_path, i, feature, output_dir): file_path
            for i, (file_path, feature) in enumerate(user_stories, start=1)
        }

        # Wait for all tasks to complete
        for future in concurrent.futures.as_completed(futures):
            file_path = futures[future]  # Get the file path associated with the future
            try:
                if not future.result():
                    logging.warning(f"Skipped processing for: {file_path} due to an error.")
            except Exception as e:
                # This is just to catch unexpected issues with threads, already logged above
                logging.error(f"Unexpected error during processing for {file_path}: {e}")
else:
    logging.warning("No user stories to process.")

# Log the total processed user stories for verification
logging.info(f"Total user stories loaded: {len(user_stories)}")
```

# pseudocode Developement agent
used to write pseudo code for given system flow

Input: system flow

Output: pseudocode

LLM: lamma3.2 / mistral / qwen2.5:1.5b / deepseek-r1:1.5b

```python
pc_agent = ReActAgent.from_tools(
    llm=llm4,
    verbose=True,
)

react_system_header_str = """
You are a pseudocode development agent.
"""
react_system_prompt = PromptTemplate(react_system_header_str)

pc_agent.update_prompts({"agent_worker:system_prompt": react_system_prompt})

prompt_dict = pc_agent.get_prompts()
for k, v in prompt_dict.items():
    print(f"Prompt: {k}\n\nValue: {v.template}")
```

```python
import os
import concurrent.futures
import logging

# Configure logging
logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(message)s")

# Define the directories
user_story_dir = "system_flows"  # Input directory for user stories
output_dir = "pseudo_codes"  # Output directory for pseudocode

# Ensure the input directory exists
if not os.path.exists(user_story_dir):
    logging.error(f"The input directory '{user_story_dir}' does not exist.")
    user_stories = []
else:
    # Get a list of all valid files in the input directory
    file_list = [
        os.path.join(user_story_dir, file)
        for file in os.listdir(user_story_dir)
        if os.path.isfile(os.path.join(user_story_dir, file))
    ]

    # Load user stories, skipping empty files
    user_stories = []
    for file_path in file_list:
        try:
            with open(file_path, "r", encoding="utf-8") as file:
                content = file.read().strip()
                if content:  # Only include non-empty files
                    user_stories.append((file_path, content))
        except Exception as e:
            logging.error(f"Failed to read file '{file_path}': {e}")

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)


def generate_and_save_pseudocode(file_path, i, feature, output_dir):
    """
    Generates pseudocode for a given user story and saves it to a file.

    Args:
        file_path (str): Path to the original user story file.
        i (int): Index of the current file (used for naming output).
        feature (str): User story content.
        output_dir (str): Directory to save the pseudocode file.
    """
    try:
        # Assuming 'pc_agent' is defined elsewhere in your code
        pseudocode = pc_agent.chat(f"""Generate pseudocode for the following system flow. Your task is to create clear, structured, and implementation-ready pseudocode for the given problem or feature. Your output must:

        Use consistent and readable syntax, following standard pseudocode conventions (e.g., IF, ELSE, FOR, WHILE, INPUT, OUTPUT, etc.).
        Include comments where necessary to explain logic or key steps.
        Focus on modularity by breaking down the logic into functions or reusable blocks when applicable.
        Cover edge cases and error-handling scenarios.
        Output only the pseudocode, strictly adhering to the task requirements. Do not include any explanatory text or additional details outside of the pseudocode. Here is the system flow: {feature}""")

        # Define a safe filename for the pseudocode
        safe_filename = f"pseudocode_{i}.txt"
        output_path = os.path.join(output_dir, safe_filename)

        # Save the pseudocode in the output directory
        with open(output_path, "w", encoding="utf-8") as file:
            file.write(str(pseudocode))  # Convert to string if necessary

        logging.info(f"Successfully processed: {os.path.basename(file_path)} -> {safe_filename}")
    except Exception as e:
        # Log the error and continue with the next user story
        logging.error(f"Error generating pseudocode for '{file_path}': {e}")


# Check if user stories are loaded correctly
if user_stories:
    logging.info(f"Processing {len(user_stories)} user stories...")
    with concurrent.futures.ThreadPoolExecutor(max_workers=5) as executor:
        # Submit tasks to the executor
        futures = [
            executor.submit(generate_and_save_pseudocode, file_path, i, feature, output_dir)
            for i, (file_path, feature) in enumerate(user_stories, start=1)
        ]

        # Collect and handle any unexpected thread errors
        for future in concurrent.futures.as_completed(futures):
            try:
                future.result()  # Raises exceptions from `generate_and_save_pseudocode`
            except Exception as e:
                # This is only for unexpected issues in the threading process
                logging.error(f"Unexpected error during pseudocode generation: {e}")
else:
    logging.warning("No user stories found to process. Ensure the input directory contains valid files.")

# Log the total number of processed user stories
logging.info(f"Total user stories processed: {len(user_stories)}")
```
